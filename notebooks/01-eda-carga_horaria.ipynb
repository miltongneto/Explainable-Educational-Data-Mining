Carga horária dos servidores

#### Imports

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from unidecode import unidecode

pd.set_option('display.max_columns', None)
```

## Carregar dados

```python
PATH = '../data/raw'
```

### Dados da carga horária do servidor

```python
# Remocao manual no arquivo "BASE_CARGA_HOR_SALA_AULA_0818_anonimizado.csv", o cabecalho estava repitido no meio do dataset. Esta linha repetida foi removida
df_carga_horaria = pd.read_csv(os.path.join(PATH, 'BASE_CARGA_HOR_SALA_AULA_0818_anonimizado.csv'), sep=';', encoding='latin-1', low_memory=False)
print(df_carga_horaria.shape)
df_carga_horaria.head()
```

### Dados dos Servidores ativos

```python
df_servidores = pd.read_csv(os.path.join(PATH, 'BASE_SERVIDORES_ATIVOS_1118.csv'), sep=';', encoding='latin-1', low_memory=False)
print(df_servidores.shape)
df_servidores.head()
```

### Filtrar apenas dados referentes aos professores

```python
df_servidores = df_servidores[df_servidores['NOMECAR_E'].str.contains('PROFESSOR', na=False)]
```

### Dados da formação do servidor

```python
df_formacao = pd.read_csv(os.path.join(PATH, 'BASE_FORMACAO_1118.csv'), sep=';', encoding='latin-1')
print(df_formacao.shape)
df_formacao.head()
```

## Agregação por Servidor

### Quantidade de disciplinas

```python
qtd_disciplinas_prof_esc = df_carga_horaria.groupby(['CPF', 'CODESC']).size()
qtd_disciplinas_prof_esc.name = 'QTD_DISCIPLINAS'
qtd_disciplinas_prof_esc = qtd_disciplinas_prof_esc.reset_index()
```

### Total de aulas

```python
total_aulas = df_carga_horaria.groupby(['CPF', 'CODESC'])['TOT_GERAL_AULA'].sum()
total_aulas.name = 'QTD_TOTAL_AULAS'
total_aulas = total_aulas.reset_index()
```

### Juntar características

```python
df_features_cpf_esc = pd.merge(qtd_disciplinas_prof_esc, total_aulas, on=['CPF', 'CODESC'])
```

```python
df_features_cpf_esc.shape
```

```python
df_jornada_features = df_features_cpf_esc.drop('CPF', axis=1).groupby('CODESC').agg(['mean', 'max'])
df_jornada_features.columns = ['JORNADA_' + c[0] + '_' + c[1] for c in df_jornada_features.columns]
df_jornada_features.reset_index(inplace=True)
df_jornada_features.fillna(0, inplace=True)
df_jornada_features.head()
```

### Junção da base de servidores com Carga Horária para obter a escola
Seguindo as instruções de https://dados.educacao.sp.gov.br/sites/all/themes/DataSEE-Theme/assets/images/editais/DesafiosNotaTecnica.pdf

```python
df_servidores[['CPF', 'UAC', 'UA_E', 'IDADE', 'ANOS_TRAB_CARGO_C', 'QTDE_ATS']]
```

```python
df_servidores_escola = pd.merge(df_carga_horaria[['CPF', 'UA', 'DI', 'CODESC']], df_servidores, left_on=['CPF', 'UA', 'DI'], right_on=['CPF', 'UA_E', 'DI'], how='inner')
```

```python
df_servidores_escola.head()
```

### Agregar por escola e computar características dos servidores 

#### Idade

```python
df_features_idade = df_servidores_escola.groupby(['CODESC'])['IDADE'].agg(['mean'])
df_features_idade.columns = 'SERVIDORES_IDADE_' + df_features_idade.columns
```

#### Tempo no cargo de classificação

```python
df_features_tempo_cargo = df_servidores_escola.groupby(['CODESC'])['ANOS_TRAB_CARGO_C'].agg(['mean'])
df_features_tempo_cargo.columns = 'SERVIDORES_TEMPO_CARGO_C_' + df_features_tempo_cargo.columns
```

#### Categoria funcional

#### Distribuição

```python
df_servidores_escola['CATEG_C'].value_counts(normalize=True)
```

#### Professores com categoria funcional diferente dentro da rede estadual
- Categoria funcional A: Docentes que ingressaram na Rede por meio de concurso público
- Categoria funcional O: Docentes que possuem vínculo temporário e excepcional com a Secretaria da Educação

```python
df_servidores_escola[df_servidores_escola['CATEG_E'].isin(['A', 'O'])].groupby('CPF')['CATEG_E'].nunique().where(lambda x: x>1).dropna()
```

```python
df_servidores_escola[df_servidores_escola['CPF'] == 154669]
```

### Agregação por escola e construção dos atributos

```python
df_features_cat_funcional = df_servidores_escola.groupby(['CODESC'])['CATEG_E'].value_counts(normalize=True).unstack().fillna(0)
df_features_cat_funcional.columns = 'SERVIDORES_CAT_FUNCIONAL_' + df_features_cat_funcional.columns
```

```python
df_features_servidores = pd.concat([df_features_idade, df_features_tempo_cargo, df_features_cat_funcional], axis=1)
df_features_servidores.reset_index(inplace=True)
df_features_servidores.head()
```

### Junção dos dados

```python
df_features = pd.merge(df_jornada_features, df_features_servidores, on='CODESC', how='inner')
print(df_features.shape)
df_features.head()
```

## Exportar dados

```python
df_features.to_csv('../data/processed/jornada.csv', index=False)
```
